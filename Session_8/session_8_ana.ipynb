1. Create a *lineplot* showing how **Study Time** varies by **Student Name**. Which student has the highest study time?


```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
    "Student": ["Ana", "Liz", "Diego", "Carla", "Enric","Ricardo"],
    "StudyTime": [4, 4, 5, 10, 9, 7]
})

sns.lineplot(data=data, x="Student", y="StudyTime", color = "blue")
plt.ylabel("Hours spend studying", fontsize = 12)
plt.xlabel("Student", fontsize = 12)
plt.title("Variation of study time per student")

plt.show()
```

```python
#Â students studying the most
maxstudytime = data["StudyTime"].max()
student = data[data["StudyTime"] == maxstudytime]["Student"].tolist()  
print(f"The student(s) that study the most is {student}")
```

2. Plot a histogram (*histplot*) of **Grade** and determine which grade range has the highest frequency of students.


```python

import pandas as pd
import seaborn as sns

data = pd.DataFrame({
    "Student": ["Ana", "Liz", "Diego", "Carla", "Enric", "Ricardo"],
    "Grade": [6, 7, 9, 10, 10, 8 ]
})

sns.histplot(data=data, x="Grade", bins=range(5, 12), color="blue")

plt.title("Frequency of Grades")
plt.xlabel("Grades", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.xticks(ticks=range(5, 11), fontsize=12)  

plt.show()
```



3. Create an ECDF plot (*ecdfplot*) for **Grade**. What is the percentage of students scoring less than 85?



```python
import pandas as pd

data = pd.DataFrame({
    "Student": ["Ana", "Liz", "Diego", "Carla", "Enric","Ricardo"],
    "Grades": [60, 75, 85, 85, 100, 80 ]
})

sns.ecdfplot(data=data, x="Grades", color="blue")

plt.title("ECDF plot for grades")
plt.xlabel("Grades", fontsize=12)
plt.ylabel("Cumulative percentage", fontsize=12)
plt.grid(True)

plt.show()
```



4. Create a *stripplot* showing **Grade** distribution for each **Course**. Which course has the most spread in grades?


```python

data = pd.DataFrame({
    "Python": [60, 75, 75, 85, 100, 90],
    "AI": [80, 70, 55, 65, 60, 90],
    "CC": [60, 75, 65, 65, 100, 75],
    "BIS": [68, 70, 79, 70, 100, 85],
})

data_long = data.melt(var_name="Course", value_name="Grade") 

sns.stripplot(data=data_long, x="Course", y="Grade", color="blue")

plt.title("Grades by course")
plt.xlabel("Course", fontsize=12)
plt.ylabel("Grades", fontsize=12)
plt.grid(True)

plt.show()
```

```python
#Â which course has the most spread
spread = data.max() - data.min()
most_spread = spread.idxmax()
print(f"The course with the most spread in grades is {most_spread} with a spread of {spread[most_spread]}")
```



5. Create a *swarmplot* to show the relationship between Gender and **Study Time**. Which gender has a higher average study time?



```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
    "Student": ["Ana", "Liz", "Diego", "Carla", "Enric","Ricardo"],
    "StudyTime": [4,4,5,10,9,7],
    "Gender": ["Female", "Female", "Male", "Female", "Male","Male"]
})

sns.swarmplot(data=data, x="Gender", y="StudyTime", color= "blue")

plt.title("Study time x gender")
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Study time (h)", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# avg study time per gender
average_study= data.groupby("Gender")["StudyTime"].mean()
print(f"The average study time per {average_study}")
```



6. Plot a *pointplot* to show the average **Grade** for each Course. Which course has the highest average grade?

```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = pd.DataFrame({
    "Python": [60, 75, 75, 85, 100, 90],
    "AI": [80, 70, 55, 65, 60, 90],
    "CC": [60, 75, 65, 65, 100, 75],
    "BIS": [68, 70, 79, 70, 100, 85],
})

data_long = data.melt(var_name="Course", value_name="Grade")

sns.pointplot(data=data_long, x="Course", y="Grade", estimator="mean", errorbar=None, color="blue")  

plt.title("Average grade by course")
plt.xlabel("Course", fontsize=12)
plt.ylabel("Average Grade", fontsize=12)

plt.show()
```

```python
# avg grade per course
average_grade = data_long.groupby("Course")["Grade"].mean()
course = average_grade.idxmax()
highest_grade= average_grade.max()

print(f"The course with the highest average grade is {course} ({highest_grade})")
```
